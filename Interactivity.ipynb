# Creating interactive widgets in notebooks

### Jupyter notebook version

The [ipywidgets](https://ipywidgets.readthedocs.io/en/stable/) package allows you to create simple interactive elements in a plot.   
These include sliders, text boxes, buttons etc, which can be used to update a graph according to user input.   
A full list of widgets in this package is available [here](https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html).

The basic idea is similar to the animation tool, so it is a good idea to look at that example first!

1. Create a basic plot with the elements you need
2. Create an update function to update the figure according to the user input.

The main tool we use in this case is the [interact](https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html) function within ipywidgets.

We'll look at a simple example, where we plot a function $\sin (w x)$, where the value of $w$ is given by a slider.

```python
# Again we need to change the backend to display dynamic content
# best to only run this cell once, at the beginning!
%matplotlib widget
```

```python
import numpy as np
import matplotlib.pyplot as plt

#interactive widgets package
import ipywidgets as ipyw
```

Basic graph:

```python
x = np.linspace(0, 2 * np.pi)
fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(1, 1, 1)
w=1.0
line, = ax.plot(x, np.sin(w*x))

ax.set_xlabel( r'$x$' , fontsize=14)
ax.set_ylabel( r'$\sin\,( %.2lf \, x)$' % w, fontsize=14)

plt.tight_layout()
plt.show()
```

Now let's add an update function!

```python
fig2 = plt.figure(figsize=(5,4))
ax = fig2.add_subplot(1, 1, 1)
w=1.0
line, = ax.plot(x, np.sin(w*x))

ax.set_xlabel( r'$x$' , fontsize=14)
ax.set_ylabel( r'$\sin\,( %.2lf \, x)$' % w, fontsize=14)

plt.tight_layout()

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    ax.set_ylabel( r'$\sin\,( %.2lf \, x)$' % w)
    fig.canvas.draw()

    
ipyw.interact(update, w=ipyw.widgets.FloatSlider(min=0.5, max=4, step=0.1, value=1.0));
```
